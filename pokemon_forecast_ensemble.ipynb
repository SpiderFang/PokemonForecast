```python
train_all = pd.read_csv("./train.csv")
test_all = pd.read_csv("./test.csv")
# print (train.shape) #(10714, 185)
print (train_all.head())
print (train_all["class"].describe()) #ground truth,ie.the predicted target
print (test_all.head())
## Correlation matrix is not suitable for this case because of columns have mixed types.
# corr_matrix = train_all.corr()
# print (corr_matrix)
# ## Heat map for correlation matrix
# fig, ax = plt.subplots(figsize=(30,30))
# sns.heatmap(corr_matrix, vmax=0.8, linewidths=0.01, linecolor="black", square=False)

## Try to use seaborn pairplot to see the relationships.
# cols = ["appearedTimeOfDay","appearedHour","appearedMinute",
#         "terrainType","closeToWater","city","continent",
#         "weather","temperature","windSpeed","windBearing","pressure","weatherIcon",
#         "population_density","urban","suburban","midurban","rural",
#         "gymDistanceKm","gymIn100m","gymIn250m","gymIn500m","gymIn1000m","gymIn2500m","gymIn5000m",
#         "pokestopDistanceKm","pokestopIn100m","pokestopIn250m","pokestopIn500m","pokestopIn1000m","pokestopIn2500m","pokestopIn5000m",
#         "class"]
# sns.pairplot(train_all[cols], size=5)
#
# appeared_cols = ["appearedTimeOfDay","appearedHour","appearedMinute","class"]
# sns.pairplot(train_all[appeared_cols], size=2.5)
# terrain_cols = ["terrainType","closeToWater","city","continent","class"]
# sns.pairplot(train_all[terrain_cols], size=2.5)
# weather_cols = ["weather","temperature","windSpeed","windBearing","pressure","weatherIcon","class"]
# sns.pairplot(train_all[weather_cols], size=2.5)
# population_cols = ["population_density","urban","suburban","midurban","rural","class"]
# sns.pairplot(train_all[population_cols], size=2.5)
# gym_cols = ["gymDistanceKm","gymIn100m","gymIn250m","gymIn500m","gymIn1000m","gymIn2500m","gymIn5000m","class"]
# sns.pairplot(train_all[gym_cols], size=2.5)
# pokestop_cols = ["pokestopDistanceKm","pokestopIn100m","pokestopIn250m","pokestopIn500m","pokestopIn1000m","pokestopIn2500m","pokestopIn5000m","class"]
# sns.pairplot(train_all[pokestop_cols], size=2.5)
# plt.tight_layout()
# plt.show()

## Carve train_data, test_data, ground_truth from train_all and test_all data frames
train_data = train_all.loc[:,"appearedTimeOfDay":"cooc_151"]
test_data = test_all.loc[:,"appearedTimeOfDay":"cooc_151"]
ground_truth = train_all["class"]
print (train_data.shape) #(10714, 183)
print (test_data.shape) #(2681, 183)
print (ground_truth) #total six classes, class:0,1,2,3,4,5
```

```python
## Check missing data
na_count = train_data.isnull().sum().sort_values(ascending=False)
print (na_count) #result: no NAs in this training dataset
```

```python
## Handle categorical and numeric datas,
## get dummies from concatenated dataset and split it to train and test back.
train_objs_num = len(train_data)
all_data = pd.concat([train_data, test_data], axis=0)
all_data_dum = pd.get_dummies(all_data)
train_data_dum = all_data_dum[:train_objs_num]
test_data_dum = all_data_dum[train_objs_num:]
print (train_data_dum.shape)
print (test_data_dum.shape)
```

```python
## Standardize
## Method 1: using preprocessing.scale
# from sklearn import preprocessing
# train_data_std = preprocessing.scale(train_data_dum)
# test_data_std = preprocessing.scale(test_data_dum)
# print (train_data_std.shape)
# print (test_data_std.shape)

## Method 2: using StandardScaler
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
train_data_std = stdsc.fit_transform(train_data_dum)
test_data_std = stdsc.transform(test_data_dum)
print (train_data_std.shape)
print (test_data_std.shape)
```

```python
# train_data_std = train_data_dum
# test_data_std = test_data_dum
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=200, criterion="gini")
model = clf.fit(train_data_std, ground_truth)
predictions = model.predict(test_data_std)
# from sklearn.metrics import mean_squared_error
# print ('RMSE is: \n', mean_squared_error(ground_truth, predictions))
```

```python
submission = pd.DataFrame()
submission['_id'] = test_all._id
submission['class'] = predictions
submission.to_csv('submission.csv', index=False)
```
